```python
# Import necessary libraries
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph (social network)
G = nx.DiGraph()

# Add nodes (users in the social network)
users = ["Alice", "Bob", "Charlie", "David", "Eve", "Frank"]
G.add_nodes_from(users)

# Add edges (connections between users)
connections = [
    ("Alice", "Bob"), ("Alice", "Charlie"), ("Bob", "David"),
    ("Charlie", "David"), ("David", "Alice"), ("Eve", "Frank"),
    ("Frank", "Eve"), ("Alice", "Eve"), ("Bob", "Frank")
]
G.add_edges_from(connections)

# Draw the graph
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)  # Position nodes with spring layout
nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="gray", node_size=2000, font_size=12)
plt.title("Social Network Graph")
plt.show()

# Compute centrality metrics
print("Centrality Metrics:")

# Degree Centrality
degree_centrality = nx.degree_centrality(G)
print("\nDegree Centrality:")
for user, score in degree_centrality.items():
    print(f"{user}: {score:.2f}")

# Betweenness Centrality
betweenness_centrality = nx.betweenness_centrality(G)
print("\nBetweenness Centrality:")
for user, score in betweenness_centrality.items():
    print(f"{user}: {score:.2f}")

# Closeness Centrality
closeness_centrality = nx.closeness_centrality(G)
print("\nCloseness Centrality:")
for user, score in closeness_centrality.items():
    print(f"{user}: {score:.2f}")

# PageRank
pagerank_scores = nx.pagerank(G)
print("\nPageRank Scores:")
for user, score in pagerank_scores.items():
    print(f"{user}: {score:.4f}")

# Visualize Degree Centrality
plt.figure(figsize=(8, 6))
node_sizes = [v * 5000 for v in degree_centrality.values()]  # Scale node size by centrality
nx.draw(G, pos, with_labels=True, node_color="lightgreen", edge_color="gray", node_size=node_sizes, font_size=12)
plt.title("Node Size by Degree Centrality")
plt.show()
```
